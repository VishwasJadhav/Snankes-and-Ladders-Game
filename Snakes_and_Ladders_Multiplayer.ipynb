```python
# Snakes and Ladders (Multiplayer) :

import random

print("WELCOME TO THE SNAKES & LADDERS GAME\n")
print("Press Enter to roll dice | Type 'EXIT' to quit")

ladders = {1: 38, 4: 14, 8: 30, 21: 42, 28: 74, 35: 79, 50: 67, 71: 92, 88: 98}
snakes = {22: 3, 32: 10, 36: 6, 48: 26, 63: 18, 82: 24, 95: 56, 97: 88, 99: 27}

# Check for correct input:
while True:
    num_players = input("Enter number of Players: ")

    if num_players.isdigit() and int(num_players) >= 1:
        num_players = int(num_players)
        break
    else:
        print("Enter a valid number (1 or more).")

# Player setup:
players = []
for i in range(num_players):
    name = input(f"Enter name of player {i + 1}:").strip().title()
    players.append({"name": name, "pos": 0})

current_player = 0

# Game loop:
while True:

    player = players[current_player]

    roll = input(f"\n{player['name']}, Press Enter to Roll a Dice > ")

    # Multiplayer Exit ----
    if roll.lower() == "exit":
        print(f"{player['name']} quit the game.")
        players.pop(current_player)

        if not players:
            print("No players left. Game over.")
            break

        current_player %= len(players)
        continue

    dice = random.randint(1, 6)
    print(f"{player['name']} rolled: {dice}")

    player["pos"] += dice

    if player["pos"] > 100:
        player["pos"] -= dice
        print(f"{player['name']} still on {player['pos']}")
        current_player = (current_player + 1) % len(players)
        continue

    print(f"{player['name']} now on {player['pos']}")

    if player["pos"] in ladders:
        player["pos"] = ladders[player["pos"]]
        print(f"{player['name']} Climbed Ladder to {player['pos']}")

    elif player["pos"] in snakes:
        player["pos"] = snakes[player["pos"]]
        print(f"{player['name']} Slid down a Snake to {player['pos']}")

    if player["pos"] == 100:
        print(f"\n{player['name']} WON THE GAME")
        break

    current_player = (current_player + 1) % len(players)
```
